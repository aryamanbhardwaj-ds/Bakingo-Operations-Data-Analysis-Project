```python
df["delivery_status"] = df["delivery_status"].str.strip().str.title()
```

```python
before = len(df)
df = df[df["net_revenue_inr"] > 0]
print(f"\nRows removed (zero/negative revenue): {before - len(df)}")
```

```python
df['order_date'] = pd.to_datetime(df['order_date'], errors='coerce')

df['order_month']   = df['order_date'].dt.to_period('M').astype(str)
df['order_quarter'] = df['order_date'].dt.to_period('Q').astype(str)
df['order_hour']    = df['order_date'].dt.hour
df['order_dayofweek'] = df['order_date'].dt.day_name()

df['profit_margin'] = ((df['profit_loss_inr'] / df['net_revenue_inr']) * 100)
df['is_loss_order'] = (df['profit_loss_inr'] < 0).astype(int)
df['is_cancelled']  = (df['delivery_status'] == "Cancelled").astype(int)
df['is_returned']   = (df['delivery_status'] == "Returned").astype(int)

print("\nNew columns added successfully")
```

```python
total_gross    = df["gross_revenue_inr"].sum()
total_discount = df["discount_amt_inr"].sum()
total_net      = df["net_revenue_inr"].sum()
total_cogs     = df["cogs_inr"].sum()
total_delivery = df["delivery_cost_inr"].sum()
total_mktg     = df["marketing_cost_inr"].sum()
total_wf       = df["workforce_cost_inr"].sum()
total_spoilage = df["spoilage_cost_inr"].sum()
total_cost     = df["total_cost_inr"].sum()
net_profit     = df["profit_loss_inr"].sum()

print(f"\n  Gross Revenue      : Rs. {total_gross:>12,.2f}")
print(f"  Discounts Given    : Rs. {total_discount:>12,.2f}  ({total_discount/total_gross*100:.1f}% of Gross)")
print(f"  Net Revenue        : Rs. {total_net:>12,.2f}")
print(f"  ----------------------------------------")
print(f"  COGS               : Rs. {total_cogs:>12,.2f}  ({total_cogs/total_net*100:.1f}% of Net)")
print(f"  Delivery Cost      : Rs. {total_delivery:>12,.2f}  ({total_delivery/total_net*100:.1f}% of Net)")
print(f"  Marketing Cost     : Rs. {total_mktg:>12,.2f}  ({total_mktg/total_net*100:.1f}% of Net)")
print(f"  Workforce Cost     : Rs. {total_wf:>12,.2f}  ({total_wf/total_net*100:.1f}% of Net)")
print(f"  Spoilage/Waste     : Rs. {total_spoilage:>12,.2f}  ({total_spoilage/total_net*100:.1f}% of Net)")
print(f"  Total Cost         : Rs. {total_cost:>12,.2f}")
print(f"  ========================================")
print(f"  NET PROFIT/(LOSS)  : Rs. {net_profit:>12,.2f}  <-- COMPANY IN LOSS")
```

```python
# Cost Breakdown

cost_names  = ["COGS", "Delivery", "Marketing", "Workforce", "Spoilage"]
cost_values = [total_cogs, total_delivery, total_mktg, total_wf, total_spoilage]

# Convert to lakhs 
cost_in_lakhs = [v / 100000 for v in cost_values]

fig, axes = plt.subplots(1, 2, figsize=(14, 5))
fig.suptitle("Figure 1 - Financial Cost Overview", fontsize=14, fontweight="bold")

#  Left Bar Graph 
axes[0].bar(cost_names, cost_in_lakhs)
axes[0].set_title("Cost Breakdown in Lakhs")
axes[0].set_xlabel("Cost Type")
axes[0].set_ylabel("Amount in Lakhs (Rs.)")

# Add value labels on top of each bar
# i = x-position (bar index)
#cost_in_lakhs[i] + 0.1 = slightly above the bar
#+ 0.1 prevents text overlap

for i in range(len(cost_names)):
    axes[0].text(i, cost_in_lakhs[i] + 0.1, str(round(cost_in_lakhs[i], 1)) + "L", ha="center")

# Right Pie Chart 
axes[1].pie(cost_values, labels=cost_names, autopct="%1.1f%%")
axes[1].set_title("Cost Share in Percentage")

plt.tight_layout()
plt.savefig("fig01_overview.png", dpi=120, bbox_inches="tight")
plt.show()
```

```python
# Monthly groupby 
monthly = df.groupby("order_month")["order_id"].count().reset_index()
monthly.columns = ["order_month", "total_orders"]

monthly["net_revenue"]    = df.groupby("order_month")["net_revenue_inr"].sum().values
monthly["total_cost"]     = df.groupby("order_month")["total_cost_inr"].sum().values
monthly["net_profit"]     = df.groupby("order_month")["profit_loss_inr"].sum().values
monthly["discount_given"] = df.groupby("order_month")["discount_amt_inr"].sum().values

# Calculate profit margin percentage
monthly["profit_margin"] = monthly["net_profit"] / monthly["net_revenue"] * 100

print("Monthly Summary Table:")
print(monthly[["order_month", "total_orders", "net_revenue", "total_cost", "net_profit", "profit_margin"]])


# Monthly Business Trends 
fig, axes = plt.subplots(2, 2, figsize=(16, 10))
fig.suptitle("Figure 2 - Monthly Business Trends", fontsize=14, fontweight="bold")

months = monthly["order_month"]

# Chart 1 - Revenue vs Cost line chart
# “This chart shows how our revenue and cost change month by month.
# By comparing these two lines, we can easily identify months where costs exceed revenue  which directly indicates loss periods.
axes[0, 0].plot(months, monthly["net_revenue"] / 10000, marker="o", label="Net Revenue")
axes[0, 0].plot(months, monthly["total_cost"]  / 10000, marker="o", label="Total Cost")
axes[0, 0].set_title("Revenue vs Cost per Month")
axes[0, 0].set_ylabel("Amount in Rs. (x10,000)")
axes[0, 0].legend()
axes[0, 0].tick_params(axis="x", rotation=40)

# Chart 2 - Monthly Profit or Loss bar chart
bar_colors = []
for v in monthly["net_profit"]:
    if v >= 0:
        bar_colors.append("green")
    else:
        bar_colors.append("red")

axes[0, 1].bar(months, monthly["net_profit"] / 1000, color=bar_colors)
axes[0, 1].axhline(0, color="black", linestyle="--")
axes[0, 1].set_title("Monthly Profit and Loss")
axes[0, 1].set_ylabel("Amount in Rs. (x1,000)")
axes[0, 1].tick_params(axis="x", rotation=40)

# Chart 3 - Total orders per month
axes[1, 0].bar(months, monthly["total_orders"])
axes[1, 0].set_title("Total Orders per Month")
axes[1, 0].set_ylabel("Number of Orders")
axes[1, 0].tick_params(axis="x", rotation=40)

# Chart 4 - Discount given per month
axes[1, 1].plot(months, monthly["discount_given"] / 1000, marker="o")
axes[1, 1].set_title("Total Discount Given per Month")
axes[1, 1].set_ylabel("Discount in Rs. (x1,000)")
axes[1, 1].tick_params(axis="x", rotation=40)

plt.tight_layout()
plt.savefig("fig02_monthly_trends.png", dpi=120, bbox_inches="tight")
plt.show()
print("Figure 2 saved.")
```

```python
# Category-wise groupby building the summary table column by column
cat_summary = df.groupby("product_category")["order_id"].count().reset_index()
cat_summary.columns = ["product_category", "total_orders"]

cat_summary["net_revenue"]    = df.groupby("product_category")["net_revenue_inr"].sum().values
cat_summary["total_cost"]     = df.groupby("product_category")["total_cost_inr"].sum().values
cat_summary["net_profit"]     = df.groupby("product_category")["profit_loss_inr"].sum().values
cat_summary["avg_discount"]   = df.groupby("product_category")["discount_pct"].mean().values
cat_summary["spoilage_total"] = df.groupby("product_category")["spoilage_cost_inr"].sum().values
cat_summary["loss_orders"]    = df.groupby("product_category")["is_loss_order"].sum().values

# Calculate profit margin and loss rate
cat_summary["profit_margin"] = cat_summary["net_profit"] / cat_summary["net_revenue"] * 100
cat_summary["loss_rate"]     = cat_summary["loss_orders"] / cat_summary["total_orders"] * 100

# Sort by profit so worst performing category appears first
cat_summary = cat_summary.sort_values("net_profit")

print("Category-wise Profitability:")
print(cat_summary[["product_category", "total_orders", "net_revenue",
                    "net_profit", "avg_discount", "loss_rate", "profit_margin"]])


# Figure 3 - Loss Driver Analysis by Category (3 charts)
fig, axes = plt.subplots(1, 3, figsize=(18, 6))
fig.suptitle("Figure 3 - Loss Driver Analysis by Product Category", fontsize=14, fontweight="bold")


# Chart 1 - Net Profit per category
bar_colors = []
for v in cat_summary["net_profit"]:
    if v >= 0:
        bar_colors.append("green")
    else:
        bar_colors.append("red")

axes[0].barh(cat_summary["product_category"], cat_summary["net_profit"] / 1000, color=bar_colors)
axes[0].axvline(0, color="black")
axes[0].set_title("Net Profit by Category")
axes[0].set_xlabel("Amount in Rs. (x1,000)")


# Chart 2 - Average discount per category highest to shortest 
cat_by_discount = cat_summary.sort_values("avg_discount", ascending=False)

axes[1].bar(cat_by_discount["product_category"], cat_by_discount["avg_discount"] * 100)
axes[1].axhline(15, color="red", linestyle="--", label="15% safe limit")
axes[1].set_title("Average Discount Percentage per Category")
axes[1].set_ylabel("Discount %")
axes[1].tick_params(axis="x", rotation=30)
axes[1].legend()


# Chart 3 - Spoilage cost per category highest to lowest 
cat_by_spoilage = cat_summary.sort_values("spoilage_total", ascending=False)

axes[2].bar(cat_by_spoilage["product_category"], cat_by_spoilage["spoilage_total"] / 1000)
axes[2].set_title("Spoilage and Waste Cost per Category")
axes[2].set_ylabel("Amount in Rs. (x1,000)")
axes[2].tick_params(axis="x", rotation=30)


plt.tight_layout()
plt.savefig("fig03_category_loss_analysis.png", dpi=120, bbox_inches="tight")
plt.show()
print("Figure 3 saved.")
```

```python
# Monthly spoilage summary 
monthly_spoilage = df.groupby("order_month")["spoilage_cost_inr"].sum().reset_index()
monthly_spoilage.columns = ["order_month", "total_spoilage"]

monthly_spoilage["net_revenue"] = df.groupby("order_month")["net_revenue_inr"].sum().values

# Spoilage rate = what % of revenue is being wasted
monthly_spoilage["spoilage_rate"] = monthly_spoilage["total_spoilage"] / monthly_spoilage["net_revenue"] * 100

print("Monthly Spoilage Rate (%):")
print(monthly_spoilage[["order_month", "total_spoilage", "spoilage_rate"]])


# Average spoilage cost by day of week
daily_spoilage = df.groupby("order_dayofweek")["spoilage_cost_inr"].mean().reset_index()

# Manually reorder days from Monday to Sunday
day_order = ["Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday", "Sunday"]
daily_spoilage["order_dayofweek"] = pd.Categorical(daily_spoilage["order_dayofweek"], categories=day_order, ordered=True)
daily_spoilage = daily_spoilage.sort_values("order_dayofweek")

# Figure 4 - Inventory and Wastage Analysis
fig, axes = plt.subplots(1, 2, figsize=(14, 5))
fig.suptitle("Figure 4 - Inventory and Wastage Analysis", fontsize=14, fontweight="bold")


# Chart 1 - Monthly spoilage rate line chart
axes[0].plot(monthly_spoilage["order_month"], monthly_spoilage["spoilage_rate"], marker="o")
axes[0].set_title("Spoilage Rate per Month")
axes[0].set_ylabel("Spoilage as % of Revenue")
axes[0].tick_params(axis="x", rotation=40)


# Chart 2 - Avg spoilage cost by day of week
bar_colors = []
for day in daily_spoilage["order_dayofweek"]:
    if day in ["Saturday", "Sunday"]:
        bar_colors.append("red")
    else:
        bar_colors.append("steelblue")

axes[1].bar(daily_spoilage["order_dayofweek"], daily_spoilage["spoilage_cost_inr"], color=bar_colors)
axes[1].set_title("Average Spoilage Cost by Day of Week")
axes[1].set_ylabel("Avg Spoilage Cost (Rs.)")
axes[1].tick_params(axis="x", rotation=30)


plt.tight_layout()
plt.savefig("fig04_wastage_inventory.png", dpi=120, bbox_inches="tight")
plt.show()
print("Figure 4 saved.")
```

```python
# Categorize delivery time into speed buckets
df["delivery_bucket"] = pd.cut(
    df["delivery_time_mins"],
    bins=[0, 45, 60, 90, 999],
    labels=["Fast <45 min", "On-Time 45-60", "Slow 60-90", "Very Late >90"]
)

# Bucket summary - column by column
bucket_summary = df.groupby("delivery_bucket", observed=True)["order_id"].count().reset_index()
bucket_summary.columns = ["delivery_bucket", "orders"]

bucket_summary["avg_rating"] = df.groupby("delivery_bucket", observed=True)["customer_rating"].mean().values
bucket_summary["net_profit"]  = df.groupby("delivery_bucket", observed=True)["profit_loss_inr"].sum().values

print("Delivery Speed vs Customer Rating:")
print(bucket_summary)


# Orders count by hour of the day
hourly = df.groupby("order_hour")["order_id"].count().reset_index()
hourly.columns = ["order_hour", "orders"]

# Find which hour has the most orders
peak_hour = hourly.loc[hourly["orders"].idxmax(), "order_hour"]
print("Peak Order Hour:", peak_hour)


# Delivery status count
status_counts = df["delivery_status"].value_counts().reset_index()
status_counts.columns = ["status", "count"]


# Figure 5 - Customer Handling and Delivery Analysis
fig, axes = plt.subplots(1, 3, figsize=(18, 5))
fig.suptitle("Figure 5 - Customer Handling and Delivery Analysis", fontsize=14, fontweight="bold")


# Chart 1 - Avg rating by delivery speed bucket
axes[0].bar(bucket_summary["delivery_bucket"], bucket_summary["avg_rating"])
axes[0].axhline(4.0, color="black", linestyle="--", label="Target Rating 4.0")
axes[0].set_title("Average Rating by Delivery Speed")
axes[0].set_ylabel("Average Customer Rating (out of 5)")
axes[0].set_ylim(0, 5.5)
axes[0].legend()
axes[0].tick_params(axis="x", rotation=15)

# Add rating value on top of each bar
for i in range(len(bucket_summary)):
    rating = round(bucket_summary["avg_rating"].iloc[i], 2)
    axes[0].text(i, rating + 0.1, str(rating), ha="center", fontsize=10)


# Chart 2 - Orders by hour of day
axes[1].bar(hourly["order_hour"], hourly["orders"])
axes[1].axvline(peak_hour, color="red", linestyle="--", label="Peak Hour: " + str(peak_hour) + ":00")
axes[1].set_title("Orders by Hour of Day")
axes[1].set_xlabel("Hour (0 to 23)")
axes[1].set_ylabel("Number of Orders")
axes[1].legend()


# Chart 3 - Delivered vs Cancelled vs Returned
axes[2].bar(status_counts["status"], status_counts["count"])
axes[2].set_title("Delivery Status Distribution")
axes[2].set_ylabel("Number of Orders")

# Add count and percentage on top of each bar
for i in range(len(status_counts)):
    count = status_counts["count"].iloc[i]
    pct   = round(count / len(df) * 100, 1)
    axes[2].text(i, count + 3, str(count) + "\n(" + str(pct) + "%)", ha="center", fontsize=9)


plt.tight_layout()
plt.savefig("fig05_customer_delivery.png", dpi=120, bbox_inches="tight")
plt.show()
print("Figure 5 saved.")
```

```python
# Customer type summary - column by column
cust_type_summary = df.groupby("customer_type")["order_id"].count().reset_index()
cust_type_summary.columns = ["customer_type", "total_orders"]

cust_type_summary["net_revenue"]    = df.groupby("customer_type")["net_revenue_inr"].sum().values
cust_type_summary["net_profit"]     = df.groupby("customer_type")["profit_loss_inr"].sum().values
cust_type_summary["avg_order_val"]  = df.groupby("customer_type")["net_revenue_inr"].mean().values
cust_type_summary["avg_rating"]     = df.groupby("customer_type")["customer_rating"].mean().values
cust_type_summary["marketing_cost"] = df.groupby("customer_type")["marketing_cost_inr"].sum().values

# CAC = total marketing cost divided by number of orders for that customer type
cust_type_summary["cac"] = cust_type_summary["marketing_cost"] / cust_type_summary["total_orders"]

print("New vs Returning Customer Comparison:")
print(cust_type_summary[["customer_type", "total_orders", "net_revenue", "avg_order_val", "avg_rating", "cac"]])


# Complaint counts - filter out rows where there is no complaint first
complaints_only = df[df["complaint_type"] != "No Complaint"]
complaint_counts = complaints_only["complaint_type"].value_counts().reset_index()
complaint_counts.columns = ["complaint", "count"]


# Figure 6 - Customer Retention and Complaint Analysis
fig, axes = plt.subplots(1, 3, figsize=(18, 6))
fig.suptitle("Figure 6 - Customer Retention and Complaint Analysis", fontsize=14, fontweight="bold")


# Chart 1 - New vs Returning customer pie chart
type_counts   = df["customer_type"].value_counts()
repeat_rate   = round(type_counts["Returning"] / len(df) * 100, 1)

axes[0].pie(type_counts, labels=type_counts.index, autopct="%1.1f%%")
axes[0].set_title("New vs Returning Customers\nRepeat Rate: " + str(repeat_rate) + "%")


# Chart 2 - Avg Order Value vs Acquisition Cost (grouped bar)
plot_data = pd.DataFrame({
    "customer_type": list(cust_type_summary["customer_type"]) * 2,
    "metric":        ["Avg Order Value"] * len(cust_type_summary) + ["Acq. Cost (CAC)"] * len(cust_type_summary),
    "value":         list(cust_type_summary["avg_order_val"]) + list(cust_type_summary["cac"])
})

sns.barplot(data=plot_data, x="customer_type", y="value", hue="metric", ax=axes[1])
axes[1].set_title("Avg Order Value vs Acquisition Cost")
axes[1].set_ylabel("Amount in Rs.")
axes[1].set_xlabel("Customer Type")


# Chart 3 - Complaint type horizontal bar
axes[2].barh(complaint_counts["complaint"], complaint_counts["count"])
axes[2].set_title("Complaint Type Distribution")
axes[2].set_xlabel("Number of Complaints")

# Add count label at the end of each bar
for i in range(len(complaint_counts)):
    count = complaint_counts["count"].iloc[i]
    axes[2].text(count + 0.5, i, str(count), va="center", fontsize=9)


plt.tight_layout()
plt.savefig("fig06_customer_retention.png", dpi=120, bbox_inches="tight")
plt.show()
print("Figure 6 saved.")
```

```python
# Workforce performance summary - column by column
workforce_perf = df.groupby("workforce_assigned")["order_id"].count().reset_index()
workforce_perf.columns = ["workforce_assigned", "total_orders"]

workforce_perf["avg_del_time"] = df.groupby("workforce_assigned")["delivery_time_mins"].mean().values
workforce_perf["avg_rating"]   = df.groupby("workforce_assigned")["customer_rating"].mean().values
workforce_perf["avg_wf_cost"]  = df.groupby("workforce_assigned")["workforce_cost_inr"].mean().values
workforce_perf["net_profit"]   = df.groupby("workforce_assigned")["profit_loss_inr"].sum().values

print("Workforce Size vs Performance:")
print(workforce_perf)


# Orders by day of week 
dow_summary = df.groupby("order_dayofweek")["order_id"].count().reset_index()
dow_summary.columns = ["order_dayofweek", "orders"]

dow_summary["avg_wf"] = df.groupby("order_dayofweek")["workforce_assigned"].mean().values

# Reorder days from Monday to Sunday so chart reads left to right correctly
day_order = ["Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday", "Sunday"]
dow_summary["order_dayofweek"] = pd.Categorical(dow_summary["order_dayofweek"], categories=day_order, ordered=True)
dow_summary = dow_summary.sort_values("order_dayofweek")


# Figure 7 - Workforce and Scheduling Analysis
fig, axes = plt.subplots(1, 2, figsize=(14, 5))
fig.suptitle("Figure 7 - Workforce and Scheduling Analysis", fontsize=14, fontweight="bold")


# Chart 1 - Avg delivery time by workforce size
staff_labels = []
for w in workforce_perf["workforce_assigned"]:
    staff_labels.append(str(w) + " Staff")

axes[0].bar(staff_labels, workforce_perf["avg_del_time"])
axes[0].axhline(60, color="red", linestyle="--", label="60 min SLA Target")
axes[0].set_title("Avg Delivery Time by Workforce Size")
axes[0].set_ylabel("Avg Delivery Time (mins)")
axes[0].legend()

# Add delivery time label on top of each bar
for i in range(len(workforce_perf)):
    del_time = round(workforce_perf["avg_del_time"].iloc[i], 1)
    axes[0].text(i, del_time + 0.5, str(del_time) + " min", ha="center", fontsize=10)


# Chart 2 - Orders per day of week
# Weekend bars get red color to highlight higher demand days
bar_colors = []
for day in dow_summary["order_dayofweek"]:
    if day in ["Saturday", "Sunday"]:
        bar_colors.append("red")
    else:
        bar_colors.append("steelblue")

axes[1].bar(dow_summary["order_dayofweek"], dow_summary["orders"], color=bar_colors)
axes[1].set_title("Order Volume by Day of Week  (Red = Weekend)")
axes[1].set_ylabel("Number of Orders")
axes[1].tick_params(axis="x", rotation=25)


plt.tight_layout()
plt.savefig("fig07_workforce_scheduling.png", dpi=120, bbox_inches="tight")
plt.show()
print("Figure 7 saved.")
```

```python
# CHANNEL PERFORMANCE 
channel_perf = df.groupby("channel").agg(
    orders       = ("order_id",           "count"),  
    net_revenue  = ("net_revenue_inr",    "sum"),    
    net_profit   = ("profit_loss_inr",    "sum"),    
    avg_discount = ("discount_pct",       "mean"),   
    avg_del_time = ("delivery_time_mins", "mean"),   
    cancellation = ("is_cancelled",       "mean"),   
).reset_index()                                      

# Profit margin = what % of revenue became profit
channel_perf["profit_margin"] = (
    channel_perf["net_profit"] / channel_perf["net_revenue"] * 100
)

# Convert cancellation from decimal (0.15) → percentage (15%)
channel_perf["cancellation"] *= 100

print("\nChannel Performance:")
print(channel_perf.to_string(index=False))  # print without row numbers


#2. OCCASION PERFORMANCE 
# Same idea but grouped by occasion (e.g. Birthday, Anniversary, etc.)
occasion_perf = df.groupby("occasion").agg(
    orders      = ("order_id",         "count"),
    net_revenue = ("net_revenue_inr",  "sum"),
    net_profit  = ("profit_loss_inr",  "sum"),
).reset_index()

occasion_perf["profit_margin"] = (
    occasion_perf["net_profit"] / occasion_perf["net_revenue"] * 100
)

# Sort so most profitable occasions appear first in the chart
occasion_perf = occasion_perf.sort_values("net_profit", ascending=False)


#3. PLOTTING 
fig, axes = plt.subplots(1, 2, figsize=(14, 5))  
fig.suptitle("Figure 8 — Channel & Occasion Performance",
             fontsize=15, fontweight="bold")

# Left chart: Profit Margin % per Channel 
margin_colors = [
    "mediumseagreen" if v >= 0 else "tomato"
    for v in channel_perf["profit_margin"]
]

axes[0].bar(channel_perf["channel"], channel_perf["profit_margin"],
            color=margin_colors, edgecolor="white")
axes[0].axhline(0, color="black", linewidth=1)  # draw a line at 0% for reference
axes[0].set_title("Profit Margin % by Sales Channel")
axes[0].set_ylabel("Profit Margin (%)")

# Add the % value as a label on top of (or below) each bar
for i, v in enumerate(channel_perf["profit_margin"]):
    label_y = v + 0.2 if v >= 0 else v - 1.0   # shift label up for +ve, down for -ve
    axes[0].text(i, label_y, f"{v:.1f}%", ha="center", fontsize=10, fontweight="bold")

# Right chart: Net Profit per Occasion 

occ_colors = [
    "mediumseagreen" if v >= 0 else "tomato"
    for v in occasion_perf["net_profit"]
]

# Divide by 1000 so the axis shows "Rs. thousands" instead of huge raw numbers
axes[1].barh(occasion_perf["occasion"], occasion_perf["net_profit"] / 1000,
             color=occ_colors, edgecolor="white")
axes[1].axvline(0, color="black", linewidth=1)  # vertical reference line at 0
axes[1].set_title("Net Profit by Occasion (Rs. Thousands)")
axes[1].set_xlabel("Rs. (x1,000)")


#4. SAVE & SHOW
plt.tight_layout()                                         
plt.savefig("fig08_channel_occasion.png", dpi=120, bbox_inches="tight")
plt.show()
print("Figure 8 saved.")
```

```python
# 1. MISSING VALUES 
print("\n[1] Missing Values per Column:")
missing = df.isnull().sum()
missing_df = missing[missing > 0]         

if len(missing_df) == 0:
    print("    No missing values found after cleaning.")
else:
    print(missing_df)


# 2. DUPLICATE ORDER IDs 
print("\n[2] Duplicate Order IDs:")
print(f"    {df.duplicated(subset='order_id').sum()} duplicates found")


#3. OUTLIER DETECTION
print("\n[3] Outlier Check on Key Columns:")

for col in ["net_revenue_inr", "profit_loss_inr", "delivery_time_mins", "customer_rating"]:
    q1    = df[col].quantile(0.25)
    q3    = df[col].quantile(0.75)
    iqr   = q3 - q1
    lower = q1 - 1.5 * iqr          
    upper = q3 + 1.5 * iqr         
    
    # Filter rows that fall outside the acceptable range
    outliers = df[(df[col] < lower) | (df[col] > upper)]
    print(f"    {col}: {len(outliers)} outliers  (acceptable range: {lower:.1f} to {upper:.1f})")


#4. DATA TYPE VALIDATION 
print("\n[4] Data Type Validation:")

expected_types = {
    "order_id":           "object",    # text / string
    "net_revenue_inr":    "float64",   # decimal number
    "customer_rating":    "float64",
    "quantity":           "int64",     # whole number
    "delivery_time_mins": "int64",
}

for col, expected in expected_types.items():
    actual = str(df[col].dtype)
    status = "OK" if actual == expected else f"MISMATCH (got {actual})"
    print(f"    {col}: Expected {expected} — {status}")


# 5. INCONSISTENT NAMING CHECK 
print("\n[5] Inconsistent Naming (delivery_status values):")
print(df["delivery_status"].value_counts())


# 6. PREPARE DATA FOR PLOTTING 
# Missing values data 
missing_counts = df.isnull().sum().reset_index()
missing_counts.columns = ["column", "missing"]
missing_counts = missing_counts[missing_counts["missing"] > 0]  # only problem columns

# Outlier counts 
outlier_counts = []
for col in ["net_revenue_inr", "profit_loss_inr", "delivery_time_mins", "customer_rating"]:
    q1  = df[col].quantile(0.25)
    q3  = df[col].quantile(0.75)
    iqr = q3 - q1
    n   = len(df[(df[col] < q1 - 1.5*iqr) | (df[col] > q3 + 1.5*iqr)])
    outlier_counts.append({"column": col, "outliers": n})

outlier_df = pd.DataFrame(outlier_counts)   


#7. PLOTTING 
fig, axes = plt.subplots(1, 2, figsize=(14, 5))
fig.suptitle("Figure 9 — Data Quality Issues", fontsize=15, fontweight="bold")

# Left chart: Missing Values
if len(missing_counts) > 0:
    axes[0].bar(missing_counts["column"], missing_counts["missing"],
                color="tomato", edgecolor="white")
    axes[0].set_title("Missing Values per Column")
    axes[0].set_ylabel("Count")
    axes[0].tick_params(axis="x", rotation=30)  
else:
    axes[0].text(
        0.5, 0.5,                       
        "No Missing Values\n(Clean Data)",
        ha="center", va="center",
        fontsize=14,
        transform=axes[0].transAxes,    
        color="mediumseagreen"
    )
    axes[0].set_title("Missing Values — Status")

# Right chart: Outlier Counts
axes[1].bar(outlier_df["column"], outlier_df["outliers"],
            color="orange", edgecolor="white")
axes[1].set_title("Outlier Count per Numeric Column")
axes[1].set_ylabel("Number of Outliers")
axes[1].tick_params(axis="x", rotation=20)

# Add the count as a label on top of each bar
for i, v in enumerate(outlier_df["outliers"]):
    axes[1].text(i, v + 0.5, str(v), ha="center", fontsize=10, fontweight="bold")


# 8. SAVE & SHOW 
plt.tight_layout()
plt.savefig("fig09_data_quality.png", dpi=120, bbox_inches="tight")
plt.show()
print("Figure 9 saved.")
```

```python
# FIGURE 11 — OPTIMIZATION SCENARIO ANALYSIS


#1. CALCULATE SAVINGS FOR EACH SCENARIO

# S1: Cap all discounts at 15%
#Finds orders where discount > 15%
#Calculates extra discount given
#Estimates how much money was lost unnecessarily
s1_benefit = (
    df[df["discount_pct"] > 0.15]["gross_revenue_inr"].sum()   
    * (df[df["discount_pct"] > 0.15]["discount_pct"].mean() - 0.15)  
)
# S2: Reduce total delivery cost by 20% 
s2_benefit = total_delivery * 0.20

# S3: Reduce total marketing cost by 25% 
s3_benefit = total_mktg * 0.25

# S4: Cut spoilage cost by 40% 
s4_benefit = total_spoilage * 0.40

# S5: Halve cancellations = recover 50% of the lost revenue from cancelled orders
s5_benefit = df[df["delivery_status"] == "Cancelled"]["net_revenue_inr"].sum() * 0.50


# 2. PRINT SCENARIO SUMMARY\
print(f"\nS1 — Cap discounts at 15%          : Rs. {s1_benefit:>10,.2f} recovery")
print(f"S2 — Reduce delivery cost by 20%   : Rs. {s2_benefit:>10,.2f} savings")
print(f"S3 — Reduce marketing cost by 25%  : Rs. {s3_benefit:>10,.2f} savings")
print(f"S4 — Cut spoilage by 40%           : Rs. {s4_benefit:>10,.2f} savings")
print(f"S5 — Halve cancellations           : Rs. {s5_benefit:>10,.2f} recovery")

# Add all savings together
total_recovery = s1_benefit + s2_benefit + s3_benefit + s4_benefit + s5_benefit

# Project what P&L would look like if all scenarios were implemented
projected_pl = net_profit + total_recovery

print(f"\n  Current Net Loss     : Rs. {net_profit:>10,.2f}")
print(f"  Total Recovery       : Rs. {total_recovery:>10,.2f}")
print(f"  PROJECTED NET P&L    : Rs. {projected_pl:>10,.2f}")


# 3. PREPARE DATA FOR PLOTTING
scenario_names  = ["S1: Cap\nDiscounts", "S2: Delivery\nCost -20%",
                   "S3: Marketing\n-25%", "S4: Spoilage\n-40%", "S5: Cut\nCancels"]
scenario_values = [s1_benefit, s2_benefit, s3_benefit, s4_benefit, s5_benefit]


#4. PLOTTING 
fig, axes = plt.subplots(1, 2, figsize=(14, 5))
fig.suptitle("Figure 11 — Optimization Scenario Analysis", fontsize=15, fontweight="bold")

#Left chart: Savings per scenario 
axes[0].bar(
    scenario_names,
    [v / 1000 for v in scenario_values],   #showing value in thousands
    color=["steelblue", "mediumseagreen", "orange", "mediumpurple", "tomato"],
    edgecolor="white"
)
axes[0].set_title("Savings per Optimization Scenario (Rs. K)")
axes[0].set_ylabel("Rs. (x1,000)")

# Add value labels on top of each bar
for i, v in enumerate(scenario_values):
    axes[0].text(i, v/1000 + 0.3, f"Rs.{v/1000:.1f}K",
                 ha="center", fontsize=9, fontweight="bold")

# Right chart: Before vs After 
before_after = ["Current\nNet Loss", "After All\nOptimizations"]
values_ba    = [net_profit / 1000, projected_pl / 1000]

colors_ba = ["tomato" if v < 0 else "mediumseagreen" for v in values_ba]

axes[1].bar(before_after, values_ba, color=colors_ba, edgecolor="white", width=0.4)
axes[1].axhline(0, color="black", linewidth=1.2)   # zero reference line
axes[1].set_title("Net P&L: Before vs After Optimization (Rs. K)")
axes[1].set_ylabel("Rs. (x1,000)")

# Add value labels  up for positive bars, down for negative bars
for i, v in enumerate(values_ba):
    label_y    = v + 1  if v >= 0 else v - 4       # avoid overlapping the bar
    label_color = "darkgreen" if v >= 0 else "darkred"
    axes[1].text(i, label_y, f"Rs.{v:.1f}K",
                 ha="center", fontsize=11, fontweight="bold", color=label_color)


# 5. SAVE & SHOW 
plt.tight_layout()
plt.savefig("fig11_optimization.png", dpi=120, bbox_inches="tight")
plt.show()
print("Figure 11 saved.")
```
